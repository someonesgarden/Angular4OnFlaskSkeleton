### 青空文庫から「吾輩は猫である」を読み込む

```python
import urllib.request
import sys
import os
import matplotlib.pyplot as plt
import MeCab
import numpy as np
from aozora import *
```

```python
aozora = Aozora("./data/wagahaiwa_nekodearu.txt")
# 文に分解する
string = '\n'.join(aozora.read())
string = re.sub('　', '', string)
strings = re.split('。(?!」)|\n', re.sub('　', '', string))
while '' in strings:  strings.remove('')  # 空行を除く
```

```python
from collections import Counter
```

```python
# 頻度順にソートして出力する
cnt = Counter(string)
print(sorted(cnt.items(), key=lambda x: x[1], reverse=True)[:50])
```

```python
#文ごとの文字数分布をヒストグラムにする
cnt = Counter([len(x) for x in strings])
print(sorted(cnt.items(), key = lambda x: [x[1], x[0]], reverse = True))
nstring = np.array([len(x) for x in strings if len(x) < 150])
plt.hist(nstring, bins=nstring.max())
plt.title("『吾輩は猫である』の文ごとの文字数分布")
plt.xlabel('文の文字数')
plt.ylabel('出現頻度')
plt.show()
```

### 単語に分解し出現頻度を数える

```python
m = MeCab.Tagger("-Ochasen")
mecablist = []
wlist = m.parse(string).splitlines()
for u in wlist:
    xlist = []
    for v in u.split():
        xlist.append(v)
    mecablist.append(xlist)
```

```python
wordbodylist = []
for u in mecablist:
    wordbodylist.append(u[0])
cnt = Counter(wordbodylist)
print(sorted(cnt.items(), key= lambda x : x[1], reverse=True)[:100])
```
