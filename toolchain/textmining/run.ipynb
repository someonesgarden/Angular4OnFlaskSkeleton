# 青空文庫から文章をダウンロードし、unzipしてルビを取り除く

```python
import urllib.request
import sys
import os
import matplotlib.pyplot as plt
import MeCab
import numpy as np
from aozora import *
```

### dataフォルダ内に文章ファイルをダウンロード

```python
 if not os.path.isdir('./data'):
        os.makedirs('./data')

urllib.request.urlretrieve('http://www.aozora.gr.jp/cards/000148/files/789_ruby_5639.zip', './data/aozora2.zip')
```

### bashコマンドでunzipする

```python
%%bash
unzip ./data/aozora2.zip
mv ./wagahaiwa_nekodearu.txt ./data/wagahaiwa_nekodearu.txt
# mv ./sodate_chikarazuyoku.txt ./data/sodate_chikarazuyoku.txt
```

### aozora.pyに定義されたCLEAN用のクラスでルビを取り除く

```python
aozora = Aozora("./data/wagahaiwa_nekodearu.txt")
# 文に分解する
string = '\n'.join(aozora.read())
string = re.sub('　', '', string)
strings = re.split('。(?!」)|\n', re.sub('　', '', string))
while '' in strings:  strings.remove('')  # 空行を除く
```

```python
for u in aozora.read():
    print(u)
```

```python
# 先頭20文について文単位で形態素解析し、名詞だけ抽出して、基本形を文ごとのリストにする
lengthlist = np.array( [len(v) for v in strings][3:23] )
print(lengthlist)
print('average', lengthlist.mean())
print('variance', lengthlist.var())
print('std-deviation', lengthlist.std())
```

```python
#for u in lengthlist: print(u)           # それぞれの文の長さを、出現順に表示
for u in sorted(lengthlist): print(u)   # それぞれの文の長さを、長さ順に表示
```

```python
plt.rcParams['font.family'] = 'IPAGothic'
fig = plt.figure()
plt.title('文の長さ（文字数）')
plt.xlabel('長さ')
plt.ylabel('頻度')
plt.hist(lengthlist, color='blue', bins=40)  # binsでヒストグラムの横軸区分数を指定 
plt.show()
```

```python
# 箱ひげ図を作る
plt.boxplot(lengthlist)
plt.xticks([1], ['吾輩'])
plt.title('箱ひげ図')
plt.grid()
plt.xlabel('文書')
plt.ylabel('文の長さの頻度')
plt.ylim([0,50])
plt.show()
```
